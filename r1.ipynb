## 2. Реализовать аппроксимацию распределений данных с помощью ядерных оценок.

```r
sleep_quality_levels <- unique(df_2$Sleep_Quality)
colors <- rainbow(length(sleep_quality_levels))


plot(NULL, xlim = range(df_2$Stress_Level_Self_Report, na.rm = TRUE), ylim = c(0, 0.15),
     xlab = "Stress Level (Self Report)", ylab = "Density", 
     main = "Density of Stress Self Report by Sleep Quality")

for (i in seq_along(sleep_quality_levels)) {
  level <- sleep_quality_levels[i]
  dt <- df_2[df_2$Sleep_Quality == level, "Stress_Level_Self_Report", drop = TRUE]
  dens <- density(dt, na.rm = TRUE)
  lines(dens, col = colors[i], lwd = 2)
  polygon(dens, col = adjustcolor(colors[i], alpha.f = 0.3), border = NA)
}

legend("topright", legend = sleep_quality_levels, fill = adjustcolor(colors, alpha.f = 0.3), 
       border = colors, bty = "n", title = "Sleep Quality")
```

```r
kernels <- c("gaussian", "epanechnikov")
bandwidths <- c(0.5, 1.0)


par(mfrow = c(length(kernels), length(bandwidths)))
for (kernel in kernels) {
  for (bw in bandwidths) {
    dt <- data$total_amount
    dens <- density(dt, kernel = kernel, bw = bw, na.rm = TRUE)
    plot(dens, main = paste("Kernel:", kernel, ", Bandwidth:", bw),
         xlab = "Total Amount", ylab = "Density", col = "blue", lwd = 2)
    polygon(dens, col = adjustcolor("blue", alpha.f = 0.3), border = NA)
  }
}
```

```r
par(mfrow=c(1,1), mar=c(5,4,4,2)) 

#png("density_histogram_plot3.png", width = 800, height = 600)
hist(data$total_amount, probability = TRUE, col=rgb(1, 1, 1, 0.6), 
     main="Density Plot and Histogram", xlab="Total Amount", 
     ylab="Density", breaks=50, border="black", xlim=c(min(data$total_amount), max(data$total_amount)), ylim=c(0,0.1))

# Добавляем KDE с гауссовым ядром
lines(density(data$total_amount, bw=1), col="blue", lwd=2) 

# Легенда
legend("topright", legend=c("Histogram", "KDE gaussian bw = 1"), 
       fill=c(rgb(1, 1, 1, 0.6), "blue"))
```

```r
normality_tests <- function(data){
    print("Normality test")
    print(length(data))
    print(ks.test(data, 'pnorm', 0, 1))
    print(shapiro.test(data))
    print(ad.test(data))
    print(cvm.test(data))
    print(sf.test(data))
    print(lillie.test(data))
}
```

```r
qq <- function(data){
    qqnorm(data)
    qqline(data, col = "red")
}
```

```r
plot_ecdf <- function(data) {
  ecdf_data <- ecdf(data)
  
  plot(ecdf_data, main = "ECDF",
       xlab = "Значение", ylab = "Эмпирическая ФР", col = "blue", verticals = TRUE)

}
```

```r
plot_envelopes <- function(data, alpha = 0.05) {
  sorted_data <- sort(data)
    
  n <- length(sorted_data)
  ecdf <- (1:n) / n

  mean_data <- mean(sorted_data)
  std_data <- sd(sorted_data)
  norm_cdf <- pnorm(sorted_data, mean = mean_data, sd = std_data)
  

  z <- qnorm(1 - alpha / 2) 
  lower_envelope <- norm_cdf - z * sqrt(norm_cdf * (1 - norm_cdf) / n)
  upper_envelope <- norm_cdf + z * sqrt(norm_cdf * (1 - norm_cdf) / n)
  

  plot(sorted_data, ecdf, type = "p", col = "blue", pch = 16,
       xlab = "Значения данных", ylab = "Функция распределения",
       main = "Метод огибающих для проверки нормальности")
  lines(sorted_data, norm_cdf, col = "red")
  lines(sorted_data, lower_envelope, col = "green", lty = 2)
  lines(sorted_data, upper_envelope, col = "green", lty = 2)
  legend("bottomright", legend = c("ЭФР", "ФР нормального распределения",
                                   paste("Огибающие (alpha =", alpha, ")")),
         col = c("blue", "red", "green"), lty = c(NA, 1, 2), pch = c(16, NA, NA))
}
```
